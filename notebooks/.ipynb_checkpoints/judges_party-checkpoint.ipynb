# Dissecting the U.S. judicial system based on party, gender, race

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42
from bs4 import BeautifulSoup
import numpy as np
import seaborn as sns
import re
%matplotlib inline
```

```python
df = pd.read_excel("selected_judges.xlsx")
df=df.rename(columns={'Race or Ethnicity': 'race','Appointing President' : 'president', 'Party of Appointing President':'party', 'Appointment Title' :'title'})
```

```python
df.groupby(by='president').party.value_counts().plot(kind='barh', width=0.5, figsize=(8,4), color='lightblue')
plt.savefig('president_party.pdf')
```

```python
df.party.value_counts().plot(kind='barh', width=0.5, figsize=(12,6), color='lightblue').set_title('Democratic vs Republican appointed judges since 1989')
plt.savefig('dem_vs_Rep.pdf')
```

```python
df.race.value_counts(ascending=True).plot(kind='barh', width=0.7, figsize=(10,5))
plt.savefig('judge_race.pdf')
```

```python
df['race'] = df.race.replace({
    'American Indian' : 'other_race',
     'African American/Hispanic': 'other_race',
     'Hispanic/White': 'other_race',
     'African American/White': 'other_race',
     'Asian American/Hispanic' : 'other_race',
    'Asian American/White': 'other_race',
    'Asian American/Pacific Islander': 'other_race' 
    
})
```

```python
df.race.value_counts(ascending=True).plot(kind='barh', width=0.7, color='lightblue', figsize=(10,5)).set_title('Race/ethnicity of U.S. judges since 1989')

plt.savefig('judge_race_modified.pdf')
```

```python
df.Gender.value_counts().plot(kind='barh')
plt.savefig('judge_gender.pdf')
```

```python
df.party.value_counts().plot(kind='pie', figsize=(15,8)).set_title("party-based distinction of judges")
plt.savefig('dem_vs_Rep_pie.pdf')
```

```python
df.Gender.value_counts().plot(kind='pie',, figsize=(15,8)).set_title("Gender-based distinction of judges")
plt.savefig('judge_gender.pdf')
```
