```python
good_df["properties.headline"] = good_df["properties.name"] + " authorized " + good_df["Orders"].map(str) + " wiretap applications"
```

```python
def nice_text(district):
    d = district.title()
    return d + " : wiretap applications concentrated in six states (California, New York, Nevada, Pennsylvania, Colorado, and New Jersey) accounted for 82 percent of all state wiretap applications"
```

```python
good_df['properties.article'] = good_df['properties.DISTRICT'].apply(lambda x: nice_text(x))
```

```python
def rand_color():
    letters = 'ABCDEF'
    import random
    from random import randint
    this_letter = random.choice(letters)
    random_num1 = randint(0,9)
    random_num2 = randint(0,9)
    hexColor = '#' + this_letter + this_letter + str(random_num1)+ str(random_num1)+ str(random_num2)+ str(random_num2)
    return hexColor
```

```python
good_df['properties.color'] = good_df.apply(lambda x: rand_color(), axis=1)
```

```python
good_df.head()
```

```python
ok_json = json.loads(good_df.to_json(orient='records'))
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(ok_json)
```
