```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import urllib
```

```python
response = requests.get('https://www.uscourts.gov/statistics-reports/analysis-reports/civil-justice-reform-act-report')
doc = BeautifulSoup(response.text)
```

```python
doc.find(class_="field field-name-body").find_all('p')[2].find('a')['href']
```

```python
url_list= []
urls = doc.find(class_="field field-name-body").find_all('p')

for url in urls[2:]:    
    url_list.append(url.find('a')['href'])
url_list
```

```python
link_list = []
for url in url_list:
    response = requests.get(f'https://www.uscourts.gov{url}')
    tables=BeautifulSoup(response.text)
    link_list.append(tables.find('tbody').find('a')['href'])
    
    
link_list
```

```python
table_list = []
for link in link_list:
    response = requests.get(f'https://www.uscourts.gov{link}')
    table=BeautifulSoup(response.text)
    row={}
    row['table'] = table.find(class_="file file-document file-application-pdf").find('a')['href']
    table_list.append(row)
table_list

df = pd.DataFrame(table_list)
df
```

```python
for urls in df['table']:
    link = 'https://www.uscourts.gov' + urls
    book_name = "/Users/suhailbhat/desktop/caseloadpdfs/" + link.split('/')[-2]
    with open(book_name, 'wb') as book:
        a = requests.get(link, stream=True)

        for block in a.iter_content(512):
            if not block:
                break

            book.write(block)
pdf_names = [url.split('/')[-2]+'.pdf' for url in df['table']]
pdf_names
```

```python
import os
def pdf_to_text(name):
    folder = "/Users/suhailbhat/desktop/caseloadpdfs/"
    input1 = folder + name
    txt_name = name.replace(".pdf",".txt")
    output1 = folder + txt_name
    os.system("pdftotext '%s' '%s'" % (input1, output1))
```

```python
for pdf_file in pdf_names:
    pdf_to_text(pdf_file)
```
